```python
import subprocess
import os
def subprocess_run(text,sp_id,save):
  # 使用子进程运行tts信号
        text = '\''+text+'\''
      # sp_id = '\''+sp_id+'\''
      # save = '\''+save+'\''
        command = ["tts", "--text", text ,'--model_name','tts_models/en/vctk/vits', '--speaker_idx', sp_id, '--out_path' ,save]
        ret = subprocess.run(command,shell=False)
        return ret

sp_id = 'p231'
save_path = '/workspace/Chater/results/'+'tts_output'+'.wav'
text = "There were a sensitivity and a beauty to her that have nothing to do with looks. She was one to be listened to, whose words were so easy to take to heart."
subprocess_run(text,sp_id,save_path)
```

```python
%cd /workspace/Chater/StyleSpeech/
import ss2
from scipy.io.wavfile import write
rate = 19000
args = dict()
args['text'] = 'They sat on the Terrace and many of the fishermen made fun of the old man and he was not angry.'
args['ref_audio'] = '/workspace/Chater/p311.wav'
args['save_path'] = './speech_out.wav'
args['lexicon_path'] = 'lexicon/librispeech-lexicon.txt'

model,config = ss2.get_model('./models/meta_stylespeech.pth')
mel = ss2.get_mel(model,config,args)
```

```python
from IPython.display import Audio
# mel = torch.load("./mel_output.pt")
import torch
# vocoder = torch.hub.load('/content/drive/MyDrive/melgan-neurips-master/', 'load_melgan',source='local')
vocoder = torch.hub.load('descriptinc/melgan-neurips', 'load_melgan')
mel = torch.tensor(mel,device=torch.device('cuda'))
print(mel.shape)
mel = mel.permute(0,2,1)
# print(mel.shape)
audio = vocoder.inverse(mel)
audio_numpy = audio[0].data.cpu().float().numpy()

Audio(audio_numpy, rate=rate)
# !unzip ./StyleSpeech-main.zip
```

```python
# 辅助制作QA数据集，记得验证json数据的格式
from nltk import word_tokenize
from nltk.corpus import stopwords
import json
words = stopwords.words('english')
for w in ['!',',','.','?','-s','-ly','','s']:
    words.append(w)
md = dict()
text = "Do you know if there's a convenience store around here?"
for t in word_tokenize(text):
    if t in words:
        md[t] = 0.5
    else:
        md[t] = 1
print(json.dumps(md))
```
